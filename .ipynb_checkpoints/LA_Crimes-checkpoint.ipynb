# Introducing LA County Crime Data

```python
import csv
import os
import datetime
import pandas as pd
import missingno as msno
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')  #suppress deprecation and incorrect usage warnings

%matplotlib inline
```

```python
path = "/Users/DonginKim/Documents/Python/Data"
os.chdir(path)
df = pd.read_csv('LA_crimes.csv')
df.columns = map(str.lower, df.columns)  #change column name to lower case
df.head()
```

```python
msno.matrix(df,figsize=(12,6))  #to visualize NAs
```

# Crime by Category

```python
category_count = pd.DataFrame(df['category'].value_counts())
category_percent = pd.DataFrame(df['category'].value_counts(normalize=True))  #to get percentile
category_df = category_count.merge(category_percent, left_index=True, right_index=True)  #merge

category_df.reset_index(inplace=True)
category_df.columns = ['category', 'count', 'percentile']
category_df
```

```python
category_new = category_df[category_df['percentile'] >= 0.08]
category_temp = category_df[category_df['percentile'] < 0.08]

category_new.loc['10'] = category_temp.sum(axis=0)
category_new.ix['10','category'] = 'OTHERS'
category_new
```

```python
#Pie Chart


plt.figure(figsize = (5,5))
colors = ['lightgrey', 'yellowgreen', 'lightcoral', 'lightskyblue', 'yellow', 'gold']
explode = (0, 0, 0, 0, 0, 0.15)

plt.pie(category_new['percentile'],colors=colors, explode=explode, shadow=True,
        autopct='%1.1f%%', startangle=90)
plt.title('Crimes in Los Angeles', y=1.05, fontsize=15) #distance from plot and size
plt.legend( loc="best", labels=category_new['category'], prop={'size': 6}) 
plt.tight_layout()
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize = (8,12))
sns.countplot(y='category', data=df, order = df['category'].value_counts().index)  #palette="Set3"
```

# Exploring Crimes by City

```python
city_count = pd.DataFrame(df['city'].value_counts())
city_percent = pd.DataFrame(df['city'].value_counts(normalize=True))  #to get percentile
city_df = city_count.merge(city_percent, left_index=True, right_index=True)  #merge

city_df.reset_index(inplace=True)
city_df.columns = ['city', 'count', 'percentile']
city_df.describe()
```

```python
city_new = city_df[city_df['count'] >= 150]
city_temp = city_df[city_df['count'] < 150]
```

```python
plt.figure(figsize = (8,12))
sns.barplot(x='count', y='city', data=city_new)  #palette="Set3"
```
